# ETL Project 

```python
#Import dependancies 

import pandas as pd
from sqlalchemy import create_engine
import numpy as np
```

### Store CSV into DataFrame

```python
# Load 'Olympics.csv' file into df DataFrame
olympics_df = pd.read_csv("../Olympics.csv", encoding="utf-8")
olympics_df
```

```python


olympics_data_df = olympics_df[['Nation', 'Population', 'Exists', 'Code','First_App', 'Apps', 'Medal.1', 
                               'Gold', 'Silver', 'Bronze', 'MostSuccessfulSport']].copy()
olympics_data_df.head()

olympics_data_df = olympics_data_df.rename(columns={"Apps": "Appearances",
                                                "Medal.1":"Medals",
                                               })

olympics_data_df
```

```python
# correct the country code for Germany 

olympics_data_df.replace(["GER"],"DEU", inplace=True)
olympics_data_df.head(75)
```

# GDP data 

```python
# Load 'GDP By Country.csv' file into df DataFrame
GDP_df = pd.read_csv("../GDP By Country.csv", encoding="utf-8")
GDP_df
```

```python

# 


GDP_data_df = GDP_df[['Country Name', 'Country Code', 'Indicator Name', '1994','1996', '1998', '2000', 
                               '2002', '2004', '2006', '2008', '2010','2012', '2014', '2016', '2018']].copy()

GDP_data_df = GDP_data_df.rename(columns={"Country Name": "Country_Name",
                                        "Country Code": "Country_Code",
                                        "Indicator Name": "Indicator_Name"
                                               })

GDP_data_df

```

```python
clean_GDP = pd.melt(GDP_data_df, id_vars=["Country_Name", "Country_Code", "Indicator_Name"], var_name="Year", value_name='GDP')
```

```python
clean_GDP[clean_GDP['Year'] == '2016' ].dropna(how='any')
```

```python
rds_connection_string = "postgres:postgres@localhost:5432/GDP_vs_Olympics"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
clean_GDP.to_sql(name='gdp', con=engine, if_exists='replace', index=False)
```

```python
olympics_data_df.to_sql(name='olympics', con=engine, if_exists='replace', index=False)
```
